## 우선 제너릭이 아닌 정수 리스트와 "정수 리스트"의 리스트를 따로 정의

클래스 메소드와 인스턴스 메소드가 같은 이름으로 한쌍으로 동작하게 대부분 만들면 된다.
(다른 방법으로 가능하면 그렇게 창의적으로 해도 정확하게만 돌면 상관없음)

`palindrome`을 제외한 클래스 메소드는 이미 완성되어 있으므로 수정하지 않는다.

 * `equals` (1점) 두 리스트에 들어있는 값이 순서와 내용이 정확히 같은지
   ```
   NodeInt l1 = ... // 1,2,3,null
   NodeInt l2 = ... // 1,2,3,null
   Node.equals(l1,l2); // 결과값이 true
   
   l1과 l2의 길이가 다르다던가 길이가 같아도 내용이 같지 않거나 내용이 같아도 순서가 다르면 false

   Node.equals(null,null); // 결과값이 true
   한쪽은 null인데 다른쪽은 null이 아니면 false
   ```
 * `append` (1점) 두 리스트를 이어붙인 새로운 리스트를 결과로 계산
   ```
   NodeInt l1 = ... // 1,2,null
   NodeInt l2 = ... // 3,4,null
   Node.append(l1,l2); // 결과는 1,2,3,4,null
   // 이렇게 하고 나서 l1과 l2는 변하지 않고 그대로
   
   Node.append(null,l2); // 결과는 그냥 l2이다
   ```
 * `reverse` (1점) 리스트에 나타나는 원소의 순서를 거꾸로 뒤집은 새로운 리스트를 계산
   ```
   NodeInt l1 = ... // 1,2,3,null
   Node.reverse(l1); // 결과는 3,2,1,null
   // 이렇게 하고 나서 l1은 그대로 변하지 않음
   ```
 * `palindrome` (1점) 리스트의 순서를 거꾸로 뒤집어도 내용이 그대로인지 검사
   ```
   이런 것들이 true가 된다
   NodeInt l0 = null
   NodeInt l1 = ... // 1,null
   NodeInt l2 = ... // 1,2,3,3,2,1,null
   NodeInt l3 = ... // 1,2,3,2,1,null
   NodeInt.palindrome(l0); // 결과가 true
   NodeInt.palindrome(l1); // 결과가 true
   NodeInt.palindrome(l2); // 결과가 true
   NodeInt.palindrome(l3); // 결과가 true
   ```
 * `tails` (1점) 주석 참고할 것
 * `flat` (1점) 주석 참고할 것

````java
class NodeInt {
    private final int value;
    private final NodeInt next;
    // 생성자
    NodeInt(int v, NodeInt n) { value=v; next=n; } 
    // 기본적인 값을 가져오기 위한 메소드
    int head() { return value; }
    NodeInt tail() { return next; }

    @Override // 리스트의 내용을 알아보기 편하게 출력되게
    public String toString() {
        return head() + ", " + tail();
    }
    
    ////////////////////////////////////////////////////////////////

    // 빈 리스트를 나타내는 null인 경우를 포함해 처리하기 위한 정적 메소드
    static int length(NodeInt xs) {
        return (xs==null)? 0: xs.length(); // null이 아니면 인스턴스 메소드 호출
    }
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    int length() {
        /*
        int count = 1;
        int new_next = tail(); 
        while ( new_next != null) {
            ++count;
            new_next = new_next.tail();
        } */
        return 1 + length( tail() ); // tail()이 null일수도 있으므로 클래스 메소드 호출
    }
    
    static boolean equals(NodeInt xs, NodeInt ys) {
        return (xs==null)? ys==null: xs.equals(ys); // null이 아니면 인스턴스 메소드 호출
    }
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    boolean equals(NodeInt ys) { // 같은 원소가 정확히 같은 순서로 나타나는 리스트인지 검사
        return ys!=null && head() == ys.head()
            && equals(tail(), ys.tail());
        /*
        if (ys==null) return false;
        
        return this.head() == ys.head() // 첫번째 것들끼리 같고
            && equals(this.tail(), ys.tail()); // 그 나머지가 equals로 검사해서 같다
            */
    }
    
    static NodeInt append(NodeInt xs, NodeInt ys) { // xs와 ys를 이어붙인 새로운 리스트
        return (xs==null)? ys: xs.append(ys);
    }
    // xs = 1,2,3,null
    // xs.head() = 1
    // xs.tail() = 2,3,null
    // ys = 4,5,null
    // 결과: 1,      2,3,4,5,null
    // append( 2,3,null   ,   4,5,null ) = 2,3,4,5,null
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    NodeInt append(NodeInt ys) { // 끝에다 주어진 리스트를 이어붙인 새로운 리스트
        return (ys==null)? this: new NodeInt(head(), append(tail(), ys) );
        /*
        if (ys==null) return this;
        // 가장 처음으로 오는 것이 this.head() 그 다음에는 this.tail()에 ys를 이어붙인 결과
        return new NodeInt(head(), append(tail(), ys) );
        */
    }
    
    static NodeInt reverse(NodeInt xs) { // xs를 뒤집은 새로운 리스트
        return (xs==null)? null: xs.reverse();
    }
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    NodeInt reverse() { // 리스트를 원소가 나열된 순서를 뒤집은 새로운 리스트
        // this = 1,2,3,null
        // this.tail() = 2,3,null
        // reverse( this.tail() ) = 3,2,null
        // 위에다가 1,null 을 이어붙이면 그러니까
        // append( 3,2,null   ,   1,null ) = 3,2,1,null
        // tail()을 뒤집은 다음에 맨 끝에다 head()로만 이루어진 리스트를 이어붙이자
        return append( reverse(tail()), new NodeInt(head(), null) );
    }    
    
    static boolean palindrome(NodeInt xs) { // xs의 순서를 뒤집어도 같은 내용의 리스트인지 검사
        return equals(xs, reverse(xs)); // 이걸 완성해보자
    } 
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    boolean palindrome() { // 순서를 뒤집어도 같은 내용의 리스트인지 검사
        // 예를 들면 null 또는 1,2,2,1,null 또는 1,2,3,2,1,null
        return palindrome(this); // 작성해보기
    }
    
    static NodeNodeInt tails(NodeInt xs) { // xs의 tail, xs의 tail의 tail, xs의 tail의 tail의 tail, ... , null
        return (xs==null)? null: xs.tails(); // null이 아니면 인스턴스 메소드 호출
    }
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    NodeNodeInt tails() {
        // this = 1,2,3,null
        // this.tails() = [2,3,null],    [3,null],[null],null
        
        // this.tail() = 2,3,null
        // tails( this.tail() ) =         [3,null],[null],null
        
        // 맨 앞에 tail() 오고 그 다음에는 tail()의 tails()가 온다
        return new NodeNodeInt( tail(), tails( tail() ) ); // 작성해보기
    }
}

class NodeNodeInt { // 정수 리스트를 원소로 하는 리스트, 즉 "정수 리스트"의 리스트이다
    private final NodeInt value;
    private final NodeNodeInt next;
    // 생성자
    NodeNodeInt(NodeInt v, NodeNodeInt n) { value=v; next=n; } 
    // 기본적인 값을 가져오기 위한 메소드
    NodeInt head() { return value; }
    NodeNodeInt tail() { return next; }

    @Override // 리스트의 내용을 알아보기 편하게 출력되게
    public String toString() {
        return "[" + head() + "], " + tail();
    }
    
    // [2,3,null],[3,null],[null],null 이런 내용의 xss가 있다면
    // NodeNodeInt.flat(xss)의 결과는 2,3,3,null가 된다
    static NodeInt flat(NodeNodeInt xss) { return (xss==null)? null: xss.flat(); }
    
    NodeInt flat() {
        return NodeInt.append( head(), flat( tail() ) );
    } // 힌트: 이것도 append를 이용하면 쉽다
}
````

````java
NodeInt l0 = null;
NodeInt l1 = new NodeInt(3, l0);
NodeInt l2 = new NodeInt(2, l1);
NodeInt l3 = new NodeInt(1, l2);

System.out.println("l3: " + l3); // 1, 2, 3, null

System.out.println(NodeInt.length(l0));
System.out.println(NodeInt.length(l1));
System.out.println(NodeInt.length(l2));
System.out.println(NodeInt.length(l3));


NodeNodeInt ll0 = null;
NodeNodeInt ll1 = new NodeNodeInt(l3, ll0);
NodeNodeInt ll2 = new NodeNodeInt(l2, ll1);
NodeNodeInt ll3 = new NodeNodeInt(l1, ll2);

System.out.println("ll3: " + ll3); // [3, null], [2, 3, null], [1, 2, 3, null], null

System.out.println("equals(l3,l3) = " + NodeInt.equals(l3,l3));
System.out.println("equals(l3,new NodeInt(7, l2)) = " + NodeInt.equals(l3,new NodeInt(7, l2)));
System.out.println("equals(l3,l2) = " + NodeInt.equals(l3,l2));
System.out.println("equals(l2,l3) = " + NodeInt.equals(l2,l3));

System.out.println("append(l2,l3) = " + NodeInt.append(l2,l3));

System.out.println("reverse(l3) = " + NodeInt.reverse(l3));

System.out.println("palindrome(l3) = " + NodeInt.palindrome(l3));
System.out.println("palindrome( l3에 reverse(l3)이어붙인것 ) = "
                   + NodeInt.palindrome(NodeInt.append(l3, NodeInt.reverse(l3))));
System.out.println("palindrome( l3에 2,1,null을 이어붙인 것 ) = "
                   + NodeInt.palindrome(NodeInt.append(l3, new NodeInt(2,new NodeInt(1,null)))));

System.out.println("tails(l3) = " + NodeInt.tails(l3));



// // flat을 제대로 정의하고 나면 이런 게 동작할 것이다.
System.out.println("NodeNodeInt.flat(ll3) = " + NodeNodeInt.flat(ll3)); // 3, 2, 3, 1, 2, 3, null
````

## 제너릭 클래스 정의

주의: `int`같은 기본타입은 그냥 `==`로 비교연산을 하면 되는데
제너릭은 기본적으로 클래스 인스턴스가 되는 타입들 즉 오브젝트만을 다루므로 `==` 대신 `equals`를 써야.

* 생성자 (1점)
* length (1점)
* equals (1점)
* append (1점)
* reverse (1점)
* tails (1점)
* palindrome (1점)
* flat (4점) - 이것은 클래스 메소드로만 정의하도록. 그게 아마 더 쉬울 것이다. 인스턴스 메소드와 함께 정의하기는 좀 까다로움.

````java
class Node<T> {
    private final T value;
    private final Node<T> next;
    
    // 생성자를 적절하게 정의하라
    Node(T v, Node<T> n) { value=v; next=n; }
    
    T head() { return value; }
    Node<T> tail() { return next; }
    
    @Override // 리스트의 내용을 알아보기 편하게 출력되게
    public String toString() {
        return
            ( (head() instanceof Node)
             ? "["+ head() + "]"
             :      head() )
            + ", " + tail();
    }

    // 제너릭 클래스에 맞게 나머지 메소드들도 정의해 보라
    // length, equals, apppend, reverse, palindrome, tails, flat
    
    
    // 빈 리스트를 나타내는 null인 경우를 포함해 처리하기 위한 정적 메소드
    static <S> int length(Node<S> xs) {
        return (xs==null)? 0: xs.length(); // null이 아니면 인스턴스 메소드 호출
    }
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    int length() {
        return 1 + length( tail() ); // tail()이 null일수도 있으므로 클래스 메소드 호출
    }
    
    static <S> boolean equals(Node<S> xs, Node<S> ys) {
        return (xs==null)? ys==null: xs.equals(ys); // null이 아니면 인스턴스 메소드 호출
    }
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    boolean equals(Node<T> ys) { // 같은 원소가 정확히 같은 순서로 나타나는 리스트인지 검사
        return ys!=null && head() == ys.head()
            && equals(tail(), ys.tail());
    }
    
    static <S> Node<S> append(Node<S> xs, Node<S> ys) { // xs와 ys를 이어붙인 새로운 리스트
        return (xs==null)? ys: xs.append(ys);
    }
    Node<T> append(Node<T> ys) { // 끝에다 주어진 리스트를 이어붙인 새로운 리스트
        return (ys==null)? this: new Node<>(head(), append(tail(), ys) );
    }
    
    static <S> Node<S> reverse(Node<S> xs) { // xs를 뒤집은 새로운 리스트
        return (xs==null)? null: xs.reverse();
    }
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    Node<T> reverse() { // 리스트를 원소가 나열된 순서를 뒤집은 새로운 리스트
        return append( reverse(tail()), new Node<>(head(), null) );
    }    
    
    static <S> boolean palindrome(Node<S> xs) { // xs의 순서를 뒤집어도 같은 내용의 리스트인지 검사
        return equals(xs, reverse(xs)); // 이걸 완성해보자
    } 
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    boolean palindrome() { // 순서를 뒤집어도 같은 내용의 리스트인지 검사
        // 예를 들면 null 또는 1,2,2,1,null 또는 1,2,3,2,1,null
        return palindrome(this); // 작성해보기
    }
    
    static <S> Node< Node<S> > tails(Node<S> xs) { // xs의 tail, xs의 tail의 tail, xs의 tail의 tail의 tail, ... , null
        return (xs==null)? null: xs.tails(); // null이 아니면 인스턴스 메소드 호출
    }
    // 위의 정적 메소드에 대응되는 인스턴스 메소드
    Node< Node<T> > tails() {
        return new Node<>( tail(), tails( tail() ) ); // 작성해보기
    }
    
    static <S> Node<S> flat(Node< Node<S> > xss) {
        return (xss==null)? null: append( xss.head(), flat( xss.tail() ) );
    }
}
````

````java
// 위에서 생성자만 제대로 정의하면 아래 코드가 에러없이 실행되어 정상적으로 출력될 것이다

Node<String> l0 = null;
Node<String> l1 = new Node<String>("c", l0);
Node<String> l2 = new Node<>("b", l1); // 생략가능
Node<String> l3 = new Node<>("a", l2);

System.out.println("l3: " + l3); // a, b, c, null

System.out.println("equals(l3,l3) = " + Node.equals(l3,l3));

Node< Node<String> > ll0 = null;
Node< Node<String> > ll1 = new Node< Node<String> >(l3, ll0);
Node< Node<String> > ll2 = new Node<>(l2, ll1); // 생략가능
Node< Node<String> > ll3 = new Node<>(l1, ll2); // 이렇게도 생략가능

System.out.println("ll3: " + ll3); // [c, null], [b, c, null], [a, b, c, null], null

// // flat을 제대로 정의하고 나면 이런 게 동작할 것이다.
System.out.println("Node.flat(ll3) = " + Node.flat(ll3)); // c, b, c, a, b, c, null

// 마찬가지로 다른 메소드에 대해서도 테스트 코드 작성
````
